```python
from tkinter import *
import math

root = Tk()
root.title("Calculator")

def My_click(num):
    v = e.get()
    e.delete(0, END)
    e.insert(0, v + str(num))

def My_crossone():
    v = e.get()
    e.delete(0, END)
    e.insert(0, v[0:len(v) - 1])

def My_clear():
    e.delete(0, END)

def My_calculation(num):
    # print("yes")
    v = str(num)
    # ['+','-','*','/']
    z = []
    if v[0] in ["+", "-", "*", "/"]:
        i = v[0]
        v = v[1:len(v):1]
    else:
        i = "+"
        pass
    for j in range(0, len(v), 1):
        if v[j] in ["+", "-", "*", "/"]:
            z += [(j, v[j])]
    if z[0][1] == '+':
        k = z[0][0]
        if i == "+":
            s1 = float(v[0:k])
        elif i == "-":
            s1 = -float(v[0:k])
        else:
            s1 = float(v[0:k])
        if z[len(z) - 1][1] == "-" and len(z) > 1:
            s2 = -float(v[k + 2:len(v)])
        else:
            s2 = float(v[k + 1:len(v)])

        e.delete(0, END)
        e.insert(0, str(s1 + s2))

    if z[0][1] == '-':
        k = z[0][0]
        if i == "+":
            s1 = float(v[0:k])
        elif i == "-":
            s1 = -float(v[0:k])
        else:
            s1 = float(v[0:k])
        if z[len(z) - 1][1] == "-" and len(z) > 1:
            s2 = -float(v[k + 2:len(v)])
        else:
            s2 = float(v[k + 1:len(v)])

        e.delete(0, END)
        e.insert(0, str(s1 - s2))

    if z[0][1] == '*':
        k = z[0][0]
        if i == "+":
            s1 = float(v[0:k])
        elif i == "-":
            s1 = -float(v[0:k])
        else:
            s1 = float(v[0:k])
        if z[len(z) - 1][1] == "-" and len(z) > 1:
            s2 = -float(v[k + 2:len(v)])
        else:
            s2 = float(v[k + 1:len(v)])

        e.delete(0, END)
        e.insert(0, str(s1 * s2))

    if z[0][1] == '/':
        k = z[0][0]
        if i == "+":
            s1 = float(v[0:k])
        elif i == "-":
            s1 = -float(v[0:k])
        else:
            s1 = float(v[0:k])
        if z[len(z) - 1][1] == "-" and len(z) > 1:
            s2 = -float(v[k + 2:len(v)])
        else:
            s2 = float(v[k + 1:len(v)])

        e.delete(0, END)
        e.insert(0, str(s1 / s2))


def My_factorial():
    v = e.get()
    My_clear()
    e.insert(0, str(math.factorial(int(v))))


def My_destroy():
    root.destroy()


def My_new_window():
    root1 = Tk()
    root1.title("Scientific calculator")
    My_destroy()

    def My_click(num, k=0):
        if k == 10:
            v = entry.get()
            entry.delete(0, END)
            entry.insert(0, v + str(num) + "(")

        else:
            v = entry.get()
            entry.delete(0, END)
            entry.insert(0, v + str(num))

    def My_crossone():
        v = entry.get()
        entry.delete(0, END)
        entry.insert(0, v[0:len(v) - 1])

    def My_clear():
        entry.delete(0, END)

    def My_calculation(num, zk = 9):
        v = str(num)
        # ['+','-','*','/']
        z = []
        if v[0] in ["+", "-", "*", "/"]:
            i = v[0]
            v = v[1:len(v):1]
        else:
            i = "+"
            pass
        for j in range(0, len(v), 1):
            if v[j] in ["+", "-", "*", "/"]:
                z += [(j, v[j])]
        if z[0][1] == '+':
            k = z[0][0]
            if i == "+":
                s1 = float(v[0:k])
            elif i == "-":
                s1 = -float(v[0:k])
            else:
                s1 = float(v[0:k])
            if z[len(z) - 1][1] == "-" and len(z) > 1:
                s2 = -float(v[k + 2:len(v)])
            else:
                s2 = float(v[k + 1:len(v)])

            if zk == 0:
                print(s1+s1 , type(s1+s2))
                return s1 + s2
            else:
                entry.delete(0, END)
                entry.insert(0, str(s1 + s2))

        if z[0][1] == '-':
            k = z[0][0]
            if i == "+":
                s1 = float(v[0:k])
            elif i == "-":
                s1 = -float(v[0:k])
            else:
                s1 = float(v[0:k])
            if z[len(z) - 1][1] == "-" and len(z) > 1:
                s2 = -float(v[k + 2:len(v)])
            else:
                s2 = float(v[k + 1:len(v)])
            if zk == 0:
                return s1 - s2
            else:
                entry.delete(0, END)
                entry.insert(0, str(s1 - s2))

        if z[0][1] == '*':
            k = z[0][0]
            if i == "+":
                s1 = float(v[0:k])
            elif i == "-":
                s1 = -float(v[0:k])
            else:
                s1 = float(v[0:k])
            if z[len(z) - 1][1] == "-" and len(z) > 1:
                s2 = -float(v[k + 2:len(v)])
            else:
                s2 = float(v[k + 1:len(v)])
            if zk == 0:
                return s1 * s2
            else:
                entry.delete(0, END)
                entry.insert(0, str(s1 * s2))

        if z[0][1] == '/':
            k = z[0][0]
            if i == "+":
                s1 = float(v[0:k])
            elif i == "-":
                s1 = -float(v[0:k])
            else:
                s1 = float(v[0:k])
            if z[len(z) - 1][1] == "-" and len(z) > 1:
                s2 = -float(v[k + 2:len(v)])
            else:
                s2 = float(v[k + 1:len(v)])
            if zk == 0:
                return s1 / s2
            else:
                entry.delete(0, END)
                entry.insert(0, str(s1 / s2))

    def My_scientific_calculation(v):
        v = str(v)
        ll=v
        My_clear()
        if v[0] == "-":
            v=v[1:len(v):1]
            i="-"
        else:
            i=""

        def calc(v, i1, i2):
            val = v[i1 + 1:i2:1]
            val = float(val)
            try:
                if "sin" in v:
                    res = math.sin(val)
                elif "cos" in v:
                    res = math.cos(val)
                elif "tan" in v:
                    res = math.tan(val)
                elif "ln" in v:
                    res = math.log(val, 2.718)
                elif "sqrt" in v:
                    res = math.sqrt(val)
                elif "1%(" in v:
                    res = 1 / (val)
                elif "e^" in v:
                    res = 2.718 ** (val)
                elif "log" in v:
                    res = math.log(val, 10)
                else:
                    pass
            except:
                entry.delete(0, END)
                entry.insert(0, "ERROR")
            if i == "-":
                res1=-res
                return res1
            else:
                return res

        k = 0
        if v.count(")") == 1:
            i1 = v.index("(")
            i2 = len(v) - 1
            k = calc(v, i1, i2)
        else:
            def work(v1):
                c = 0
                l = []
                l1 = []
                for j in v1:
                    if "(" == j:
                        l = l + [c]
                        c += 1
                    elif j == ")":
                        l1 = [c] + l1
                        c += 1
                    else:
                        c += 1
                return l, l1

            m = v.count("(")
            v1 = "(" + v + ")"
            v2 = ""
            for j in range(m, 0, -1):
                k1, k2 = work(v1)
                v2 = v1[k1[-2] + 1:k2[-2]:1]
                k = calc(v2, v2.index("("), len(v2) - 1)
                v1 = v1[0:k1[-2]:1] + "(" + str(k) + ")" + v1[k2[-2]:len(v1) - 1:1]
        if k==0:
            return ll
        else:
            return str(k)

    def assign(v ,zk = 9):
        if v[0] == "-":
            v=v[1:len(v):1]
            l="-"
        else:
            l=""
        c = 0
        x = []
        typ = []
        x1 = []
        l1 = []
        for i in range(0, len(v), 1):
            if v[i] in ["+", "-", "*", "/"]:
                c = c + 1
                x = x + [i]
                typ += [v[i]]
            if v[i] == ")":
                l1 += [i]
        c1 = 0
        pos = 0
        lpos = []
        for j in ["sin", "cos", "tan", "sqrt", "1%(", "ln", "e^", "log"]:
            if j in v:
                n = v.count(j)
                c1 += n
                x1 += [c1]
                pos = v.find(j)
                lpos += [pos]

        lpos.sort()
        if c >= 1:
            if c1 > 1:
                g = My_scientific_calculation(v[0:x[0]:1])
                g1 = My_scientific_calculation(v[x[0] + 1:len(v):1])
                g2 = g + typ[0] + g1
                if l == "-":
                    My_calculation("-" + g2 , zk)
                else:
                    My_calculation(g2, zk)
            elif c1 == 1:
                t = v[lpos[0]:l1[len(l1) - 1]+1:1]
                g = My_scientific_calculation(t)
                g1 = v[0:v.index(t[0]):1]+v[v.index(t[0])+len(t)+1:len(v):1]
                g2=""
                for i in g1:
                    if i in ["+","-","*","/"]:
                        pass
                    else:
                        g2+=i
                if v.index(t)>v.index(g2):
                    if l == "-":
                        My_calculation("-" + g2 + typ[0] + g ,zk)
                    else:
                        My_calculation(g2 + typ[0] + g , zk)
                else:
                    if l == "-":
                        My_calculation("-" + g + typ[0] + g2 , zk)
                    else:
                        My_calculation(g + typ[0] + g2 , zk)
            elif c1 == 0:
                if l == "-":
                    My_calculation("-" + v , zk)
                else:
                    My_calculation(v, zk)
            else:
                pass
        if c == 0:
            if l == "-":
                g = My_scientific_calculation("-" + v)
            else:
                g = My_scientific_calculation(v)

            entry.delete(0, END)
            entry.insert(0, g)
    def display(l2):
        entry.delete(0, END)
        entry.insert(0, l2)

    bt1 = Button(root1, text="1", command=lambda: My_click(1), height=5, width=10)
    bt2 = Button(root1, text="2", command=lambda: My_click(2), height=5, width=10)
    bt3 = Button(root1, text="3", command=lambda: My_click(3), height=5, width=10)
    bt4 = Button(root1, text="4", command=lambda: My_click(4), height=5, width=10)
    bt5 = Button(root1, text="5", command=lambda: My_click(5), height=5, width=10)
    bt6 = Button(root1, text="6", command=lambda: My_click(6), height=5, width=10)
    bt7 = Button(root1, text="7", command=lambda: My_click(7), height=5, width=10)
    bt8 = Button(root1, text="8", command=lambda: My_click(8), height=5, width=10)
    bt9 = Button(root1, text="9", command=lambda: My_click(9), height=5, width=10)
    bt0 = Button(root1, text="0", command=lambda: My_click(0), height=5, width=10)
    btdot = Button(root1, text=".", command=lambda: My_click("."), height=5, width=10)
    clear = Button(root1, text="Clear", command=lambda: My_clear(), height=5, width=10)
    crossone = Button(root1, text="c1", command=lambda: My_crossone(), height=5, width=10)
    firstcomma = Button(root1, text="(", command=lambda: My_click("("), height=5, width=10)
    endcomma = Button(root1, text=")", command=lambda: My_click(")"), height=5, width=10)
    equal = Button(root1, text="=", command=lambda: assign(entry.get()), height=5, width=10)
    entry = Entry(root1, text="ENTER", borderwidth=10, width=50)

    sin = Button(root1, text="sin", command=lambda: My_click("sin", 10), height=5, width=10, bg="gray")
    cos = Button(root1, text="cos", command=lambda: My_click("cos", 10), height=5, width=10, bg="gray")
    tan = Button(root1, text="tan", command=lambda: My_click("tan", 10), height=5, width=10, bg="gray")
    ln = Button(root1, text="ln", command=lambda: My_click("ln", 10), height=5, width=10, bg="gray")
    log = Button(root1, text="log", command=lambda: My_click("log", 10), height=5, width=10, bg="gray")
    sqrt = Button(root1, text="sqrt", command=lambda: My_click("sqrt", 10), height=5, width=10, bg="gray")
    e = Button(root1, text="e", command = lambda: My_click("e^", 10), height=5, width=10, bg="gray")
    reciprocal = Button(root1, text="1/x", command=lambda: My_click("1%", 10), height=5, width=10, bg="gray")
    add = Button(root1, text="+", command=lambda: My_click("+", 1), height=5, width=10, bg="blue")
    sub = Button(root1, text="-", command=lambda: My_click("-", 1), height=5, width=10, bg="red")
    mul = Button(root1, text="*", command=lambda: My_click("*", 1), height=5, width=10, bg="pink")
    div = Button(root1, text="/", command=lambda: My_click("/", 1), height=5, width=10, bg="yellow")

    entry.grid(row=0, column=0, columnspan=7)
    bt7.grid(row=1, column=0)
    bt8.grid(row=1, column=1)
    bt9.grid(row=1, column=2)
    bt4.grid(row=2, column=0)
    bt5.grid(row=2, column=1)
    bt6.grid(row=2, column=2)
    bt1.grid(row=3, column=0)
    bt2.grid(row=3, column=1)
    bt3.grid(row=3, column=2)
    bt0.grid(row=4, column=1)
    btdot.grid(row=5, column=2)
    equal.grid(row=5, column=3)
    clear.grid(row=5, column=1)
    crossone.grid(row=4, column=2)
    endcomma.grid(row=4, column=0)
    firstcomma.grid(row=5, column=0)

    sin.grid(row=4, column=5)
    cos.grid(row=4, column=6)
    tan.grid(row=1, column=6)
    ln.grid(row=1, column=5)
    sqrt.grid(row=3, column=6)
    reciprocal.grid(row=3, column=5)
    add.grid(row=1, column=3)
    sub.grid(row=2, column=3)
    mul.grid(row=3, column=3)
    div.grid(row=4, column=3)
    log.grid(row=2, column=6)
    e.grid(row=2, column=5)

    root1.mainloop()


bt1 = Button(root, text="1", command=lambda: My_click(1), height=5, width=10)
bt2 = Button(root, text="2", command=lambda: My_click(2), height=5, width=10)
bt3 = Button(root, text="3", command=lambda: My_click(3), height=5, width=10)
bt4 = Button(root, text="4", command=lambda: My_click(4), height=5, width=10)
bt5 = Button(root, text="5", command=lambda: My_click(5), height=5, width=10)
bt6 = Button(root, text="6", command=lambda: My_click(6), height=5, width=10)
bt7 = Button(root, text="7", command=lambda: My_click(7), height=5, width=10)
bt8 = Button(root, text="8", command=lambda: My_click(8), height=5, width=10)
bt9 = Button(root, text="9", command=lambda: My_click(9), height=5, width=10)
bt0 = Button(root, text="0", command=lambda: My_click(0), height=5, width=10)
btdot = Button(root, text=".", command=lambda: My_click("."), height=5, width=10)
e = Entry(root, text="ENTER", borderwidth=10, width=30)
clear = Button(root, text="Clear", command=lambda: My_clear(), height=5, width=10)
crossone = Button(root, text="c1", command=lambda: My_crossone(), height=5, width=10)
sccal = Button(root, text="Fx", command=lambda: My_new_window(), height=5, width=10)
# calculation
btadd = Button(root, text="+", command=lambda: My_click("+"), height=5, width=10, bg="blue")
btsub = Button(root, text="-", command=lambda: My_click("-"), height=5, width=10, bg="red")
btmul = Button(root, text="*", command=lambda: My_click("*"), height=5, width=10, bg="pink")
btdiv = Button(root, text="/", command=lambda: My_click("/"), height=5, width=10, bg="yellow")
btequal = Button(root, text="=", command=lambda: My_calculation(e.get()), height=5, width=10, bg="white")
btfactorial = Button(root, text="!", command=lambda: My_factorial(), height=5, width=10, bg="white")

e.grid(row=0, column=0, columnspan=4)

btadd.grid(row=1, column=3)
btsub.grid(row=2, column=3)
btmul.grid(row=3, column=3)
btdiv.grid(row=4, column=3)
btequal.grid(row=5, column=3)

clear.grid(row=5, column=1)
crossone.grid(row=4, column=2)
btdot.grid(row=4, column=0)
bt7.grid(row=1, column=0)
bt8.grid(row=1, column=1)
bt9.grid(row=1, column=2)
bt4.grid(row=2, column=0)
bt5.grid(row=2, column=1)
bt6.grid(row=2, column=2)
bt1.grid(row=3, column=0)
bt2.grid(row=3, column=1)
bt3.grid(row=3, column=2)
bt0.grid(row=4, column=1)
btfactorial.grid(row=5, column=0)
sccal.grid(row=5, column=2)

root.mainloop()
```
